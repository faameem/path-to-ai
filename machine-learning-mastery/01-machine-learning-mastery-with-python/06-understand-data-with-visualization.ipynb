```python
from pandas import read_csv
```

```python
_uri = 'https://archive.ics.uci.edu/ml/machine-learning-databases/pima-indians-diabetes/pima-indians-diabetes.data'
```

```python
_col_names = ['preg','plas','pres','skin','test','mass','pedi','age','class']
```

```python
_data = read_csv(_uri, names=_col_names)
```

```python
# Univariate Plots

# 1. Histograms.
# 2. Density Plots.
# 3. Box and Whisker Plots.
```

```python
# Histograms (univariate).

# - Histograms group data into bins and provide you a count of the number of observations in each bin. 
# - From the shape of the bins you can quickly get a feeling for whether an attribute is Gaussian, skewed 
# or even has an exponential distribution. 
# - It can also help you see possible outliers.
```

```python
# pyplot tutorial
# https://matplotlib.org/users/pyplot_tutorial.html
from matplotlib import pyplot
```

```python
_data.hist()
```

```python
pyplot.subplots_adjust(top=1.00, bottom=0.00, left=0.00, right=1.00, hspace=0.75,wspace=0.75)
```

```python
pyplot.show()
```

```python
# We can see: 
# 1. Age, pedi and test may have an exponential distribution. 
# 2. Mass, pres and plas may have a Gaussian or nearly Gaussian distribution. 
# This is interesting because many machine learning techniques assume a Gaussian univariate distribution 
# on the input variables.
```

```python
# Density Plots

# Density plots look like an abstracted histogram with a smooth curve drawn through the top of each bin, 
# much like your eye tried to do with the histograms.
```

```python
_data.plot(kind='density', subplots=True, layout=(3,3), sharex=False)
```

```python
pyplot.subplots_adjust(top=1.00, bottom=0.00, left=0.00, right=1.00, hspace=0.75,wspace=0.75)
```

```python
pyplot.show()
```

```python
# Box and Whisker Plots (boxplots)

# The whiskers give an idea of the spread of the data and dots outside of the whiskers show candidate 
# outlier values (values that are 1.5 times greater than the size of spread of the middle 50% of the data).
```

```python
_data.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False)
```

```python
pyplot.subplots_adjust(top=1.00, bottom=0.00, left=0.00, right=1.00, hspace=0.75,wspace=0.75)
```

```python
pyplot.show()
```

```python
# Spread of the attributes seem to be quite different from each other.
# age, test, skin appear quite skewed towards smaller values
```

```python
# Multivariate Plots

# 1. Correlation Matrix Plot.
# 2. Scatter Plot.
```

```python
# Correlation Matrix Plot
```

```python
_corr = _data.corr()
```

```python
_figure = pyplot.figure(figsize=(10,10))
```

```python
_ax = _figure.add_subplot(111)
```

```python
_cax = _ax.matshow(_corr, vmin=-1, vmax=1)
```

```python
_figure.colorbar(_cax)
```

```python
import numpy
```

```python
_ticks = numpy.arange(0,9,1)
```

```python
_ax.set_xticks(_ticks)
```

```python
_ax.set_yticks(_ticks)
```

```python
_ax.set_xticklabels(_col_names)
```

```python
_ax.set_yticklabels(_col_names)
```

```python
pyplot.show()
```

```python
# Scatter Plot Matrix
```

```python
# You can create a scatter plot for each pair of attributes in your data. 
# Drawing all these scatter plots together is called a scatter plot matrix. 

# Scatter plots are useful for spotting structured relationships between variables, 
# like whether you could summarize the relationship between two variables with a line. 

# Attributes with structured relationships may also be correlated and good candidates for removal from your dataset.
```

```python
from pandas.tools.plotting import scatter_matrix
```

```python
scatter_matrix(_data)
```

```python
pyplot.subplots_adjust(top=1.00, bottom=0.00, left=0.00, right=1.00, hspace=0.75,wspace=0.75)
```

```python
pyplot.show()
```

```python
# Like the Correlation Matrix Plot above, the scatter plot matrix is symmetrical. 
# This is useful to look at the pairwise relationships from different perspectives
```
